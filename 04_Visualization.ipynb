### Data Visualization using Pandas and Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\reshm\Documents\My Jupyter Repo\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#df.plot.bar()
#df.plot.bar(stacked = True)
#df.plot.barh()
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Orange')
```

```python
df.plot.hist(bins = 5)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```
