### Group By

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\reshm\Documents\My Jupyter Repo\movierating.csv")
df
```

```python
group_by_frame = df.groupby('Genre')
```

```python
group_by_frame.mean(numeric_only=True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.min()
```

```python
group_by_frame.max()
```

```python
group_by_frame.sum()
```

```python
group_by_frame.agg({'Genre': ['max','min','count','sum']})
```
