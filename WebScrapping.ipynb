```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
```

```python
url = 'https://afltables.com/afl/stats/teams/geelong/2025_gbg.html'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#print(soup.prettify())

print(soup)
```

```python
total_tables = soup.find_all('table')

print(f"Total tables are: {len(total_tables)}")
```

```python
stats_cat_dr = soup.find_all('th', attrs ={'colspan':True})

print(stats_cat_dr)
```

```python
stats_cat = [title.text for title in stats_cat_dr]
print(stats_cat)
```

```python
stats_cat_disposal = soup.find_all('table')[0] #Disposal

stats_cat_disposal_thead = stats_cat_disposal.find('thead')
#print(stats_cat_header_thead)
stats_cat_disposal_th = stats_cat_disposal_thead.find_all(lambda tag: tag.name == 'th' and not tag.has_attr('colspan'))

sc_disposal_hr = [header.text.strip() for header in stats_cat_disposal_th]

print(sc_disposal_hr)
```

```python
import pandas as pd
```

```python
sc_disposal_df = pd.DataFrame(columns = sc_disposal_hr)
sc_disposal_df
```

```python
column_data = stats_cat_disposal.find_all('tr')

for row in column_data[2:-2]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    length = len(sc_disposal_df)
    sc_disposal_df.loc[length] = individual_row_data
    
```
